```python
from llama_cpp import Llama
from huggingface_hub import hf_hub_download
import os

# Define the model ID and filename
model_id = "TheBloke/Llama-2-7B-Chat-GGUF"
model_basename = "llama-2-7b-chat.Q4_K_M.gguf"
local_model_directory = "."
local_model_path = os.path.join(local_model_directory, model_basename)

# Check if the model already exists to avoid re-downloading
if not os.path.exists(local_model_path):
    print(f"Downloading model '{model_basename}' from '{model_id}' to '{local_model_directory}'...")
    hf_hub_download(
        repo_id=model_id,
        filename=model_basename,
        local_dir=local_model_directory
    )
    print("Download complete.")
else:
    print(f"Model '{model_basename}' already exists in '{local_model_directory}', skipping download.")

llm = Llama(model_path=local_model_path, n_ctx=4096)
```

```python
def chunk_text(text, chunk_size=500, overlap=100):
  chunks=[]
  start = 0

  while start < len(text):
    end = start + chunk_size
    chunk = text[start:end]
    chunks.append(chunk)
    start += chunk_size - overlap
  return chunks
```

```python
from pypdf import PdfReader

def load_pdf(path):
  reader = PdfReader(path)
  pages = []

  for i, page in enumerate(reader.pages):
    text = page.extract_text() or ""
    pages.append({
        "text" : text,
        "metadata" : {
            "source": path,
            "page": i+1
        }
    })
  return pages
```

```python
import chromadb

chroma_client = chromadb.Client()

collection = chroma_client.get_or_create_collection(
    name = "pdf_rag",
    metadata = {"hnsw:space": "cosine"}
)
```

```python
import os, uuid

def ingest_pdfs(folder="/content/sample_data/data"):
  for file in os.listdir(folder):
    if not file.endswith(".pdf"):
      continue
    full_path = os.path.join(folder,file)
    pages = load_pdf(full_path)


    for page in pages:
      chunks = chunk_text(page["text"])

      for chunk in chunks:
        id_ = str(uuid.uuid4())
        collection.add(
            ids = [id_],
            documents = [chunk],
            metadatas = [page["metadata"]]
        )

```

```python
def retrieve(query, k=5):
    results = collection.query(
        query_texts=[query],
        n_results=k,
    )
    docs = results["documents"][0]
    metas = results["metadatas"][0]
    return list(zip(docs, metas))
```

```python
def rag(query):
    retrieved = retrieve(query)

    context = ""
    for doc, meta in retrieved:
        context += f"[PDF: {meta['source']} | Page: {meta.get('page','N/A')}]\n"
        context += doc + "\n\n"

    prompt = f"""
You are an expert assistant. Use ONLY the context below to answer.

CONTEXT:
{context}

QUESTION:
{query}

ANSWER:
"""

    out = llm(prompt, max_tokens=300)
    return out["choices"][0]["text"]
```

```python
ingest_pdfs()

```

```python
print(rag("Explain the influence of paganism and Christianity from the PDF."))
```
